```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
s = nlp('This is a U.S. country. This is test sentence')
```

```python
type(s)
```

# STEMMING VS LEMMATISATION

```python
#tokenization
for word in s:
    print(word)
```

```python
#sentence spliting
for sentence in s.sents:
    print(sentence)
```

```python
s = nlp('This class is going and students are attending the class')
```

```python
s
```

```python
for word in s:
    print(f'{word}-{word.lemma_}')
```

```python
for word in s:
    print(f'{word} --> {word.pos_}')
```

```python
s = nlp('Apples are present in cafeteria of Apple company')
```

```python
from spacy import displacy
displacy.serve(s, style="ent")
```
